```python
import pandas as pd
import numpy as py
import matplotlib.pyplot as plt
import seaborn as sns 
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

```python
features = df.drop(columns=['Wine'])
scaler = StandardScaler()
scaled_features = scaler.fit_transform(features)
```

```python
features
```

```python
pca = PCA()
principal_components = pca.fit_transform(scaled_features)

explained_variance = pca.explained_variance_ratio_

plt.figure(figsize=(10, 5))
plt.plot(range(1, len(explained_variance) + 1), py.cumsum(explained_variance), marker='o', linestyle='--', color='b')
plt.xlabel("Number of Principal Components")
plt.ylabel("Cumulative Explained Variance")
plt.title("Scree Plot")
plt.grid(True)
plt.show()
```

```python
cumulative_variance = py.cumsum(explained_variance)
num_components = py.argmax(cumulative_variance >= 0.90) + 1

num_components
```

```python
print(f"Total number of principal components:-{pca.n_components_}")
```

```python
pca_df = pd.DataFrame(principal_components, columns = [f'PC{i+1}' for i in range(pca.n_components_)])
print(pca_df.head())
```

```python
n_components_reduction = 9
pca = PCA(n_components = n_components_reduction)
principal_components_after_reduction = pca.fit_transform(scaled_features)
print(f"Reduced number of components {n_components_reduction}")
```
